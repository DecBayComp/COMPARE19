This notebook describes the steps to plug into TRamWAy's `infer` logic some code that gives quantiles of the displacement amplitude, per default the maximum amplitude.

The code to be plugged-in is the [dr.py](../0. Installation/dr.py) file.
It has to be copied into the active copy of the TRamWAy project (in the Python path).

To this aim, let us use the git version of TRamWAy instead:

    pip3 uninstall tramway
    git clone https://github.com/DecBayComp/TRamWAy
    cd TRamWAy
    pip3 install --user -e .
    
The above sequence of commands assumes that TRamWAy has been previously installed a proper way (e.g. from PyPI).
Otherwise the full procedure is the following:

    git clone https://github.com/DecBayComp/TRamWAy
    cd TRamWAy
    pip3 install --user -r requirements.txt
    pip3 install --user -e .[animate]
    
Eventually, copy the [dr.py](../0. Installation/dr.py) file into this local copy:

    cp /path/to/dr.py tramway/inference/
    

```python
# demonstrate the 'dr' module

%matplotlib inline
from tramway.helper import *

# input rwa file and analysis tree
analysis_file = 'transient_sinks.rwa'
analysis_tree = load_rwa(analysis_file)

# pick and choose an existing segmentation
segmentation_label = 'hexagonal mesh'

# run 'dr'
infer(analysis_tree, 'dr', input_label=segmentation_label, output_label='max displacement')
infer(analysis_tree, 'dr', input_label=segmentation_label, output_label='large displacement', quantile=.9)

# draw a map
map_plot(analysis_tree, label=[segmentation_label,   'max displacement'], unit='$\mu m$', figsize=(12,9))
map_plot(analysis_tree, label=[segmentation_label, 'large displacement'], unit='$\mu m$', figsize=(12,9))
```
