This notebook shows a series of different tessellations grown on spatially structured data with default parameters.

```python
# initialize

%matplotlib inline
import os.path
from tramway.helper import *

trajectory_file = 'glycine_receptor.trxyt'

cell_plot_kwargs = dict(voronoi=dict(centroid_style=None), figsize = (16,12))

analysis_tree = Analyses(load_xyt(trajectory_file))
```

```python
# random cell centers

random_label = 'random'
tessellate(analysis_tree, 'random', output_label=random_label)
cell_plot(analysis_tree, label=random_label, **cell_plot_kwargs)
```

```python
# regular square grid

square_label = 'square'
tessellate(analysis_tree, 'grid', output_label=square_label)
cell_plot(analysis_tree, label=square_label, **cell_plot_kwargs)
```

```python
# quad-tree

quadtree_label = 'quadtree'
tessellate(analysis_tree, 'kdtree', output_label=quadtree_label)
cell_plot(analysis_tree, label=quadtree_label, **cell_plot_kwargs)
```

```python
# regular hexagonal grid

hexagon_label = 'hexagon'
tessellate(analysis_tree, 'hexagon', output_label=hexagon_label)
cell_plot(analysis_tree, label=hexagon_label, **cell_plot_kwargs)
```

```python
# k-means

kmeans_label = 'kmeans'
tessellate(analysis_tree, 'kmeans', output_label=kmeans_label, prune=True)
cell_plot(analysis_tree, label=kmeans_label, **cell_plot_kwargs)
```

```python
# growing when required

gwr_label = 'gwr'
tessellate(analysis_tree, 'gwr', output_label=gwr_label)
cell_plot(analysis_tree, label=gwr_label, **cell_plot_kwargs)
```
